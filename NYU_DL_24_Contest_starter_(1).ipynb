```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 2048 # Choose any
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.
```

## Load model and wrap with LoRA adapters

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

## Competition dataset

```python
# download and load competition dataset

from datasets import load_dataset
dataset = load_dataset("ad6398/nyu-dl-teach-maths-comp")
# print and see dataset
print(dataset)
```

```python
print(dataset['train'].to_pandas().head())
```

```python
prompt = """You are an expert math grader. Your task is to judge whether the explanation is reasonable for question. Respond with 'True' if the explanation is reasonable and 'False' if it is not.

Below are the Question and Explanation.





### Question:
{}

### Explainaition:
{}

### Output:
{}"""

EOS_TOKEN = tokenizer.eos_token # Must add EOS_TOKEN
def formatting_prompts_func(examples):
    question = examples["question"]
    # ans       = examples["answer"]
    output      = examples["is_correct"]
    explainaition = examples["solution"]
    texts = []
    for instruction, explainaition, output in zip(question,  explainaition, output):
        # Must add EOS_TOKEN, otherwise your generation will go on forever!
        text = prompt.format(instruction,explainaition,output) + EOS_TOKEN
        texts.append(text)
    return { "text" : texts, }
```

```python
# Process the training dataset and generate prompt for each datapoint

train_dataset = dataset['train'].map(formatting_prompts_func, batched = True,)
```

```python
#print a smaple training example
train_dataset['text'][0]
```

```python
train_dataset = train_dataset.shuffle()
```

## SFT

```python
from trl import SFTTrainer
from transformers import TrainingArguments
from unsloth import is_bfloat16_supported

training_args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 10,
        # num_train_epochs = 10, # Set this for 1 full training run.
        max_steps = 1000,
        learning_rate = 1e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 10,
        optim = "adamw_8bit",
        weight_decay = 0.005,
        lr_scheduler_type = "linear",
        # seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    )

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = train_dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    dataset_num_proc = 1,
    packing = False, # Can make training 5x faster for short sequences.
    args = training_args
)
```

```python
trainer_stats = trainer.train()
```

```python
import csv
from tqdm import tqdm

def evaluate_model_on_test_set(model, tokenizer, test_dataset, prompt, device="cuda"):
    correct_predictions = 0
    total_samples = len(test_dataset)
    predictions = []

    for i in tqdm(range(total_samples)):

        sample_ques = test_dataset['question'][i]
        # sample_ans = test_dataset['answer'][i]
        sample_sol = test_dataset['solution'][i]
        expected_output = "True" if test_dataset['is_correct'][i] else "False"


        input_prompt = prompt.format(sample_ques, sample_sol, "")
        inputs = tokenizer([input_prompt], return_tensors="pt").to(device)
        input_shape = inputs['input_ids'].shape
        input_token_len = input_shape[1]  # Length of input tokens


        outputs = model.generate(**inputs, max_new_tokens=64, use_cache=True)
        response = tokenizer.batch_decode([outputs[0][input_token_len:]], skip_special_tokens=True)
        generated_output = response[0].strip()  # Get the generated answer and strip any whitespace


        is_correct = generated_output == expected_output
        if is_correct:
            correct_predictions += 1



        predictions.append({'ID': i, 'is_correct': generated_output})


    accuracy = correct_predictions / total_samples
    print(f"Model Accuracy on Test Set: {accuracy * 100:.2f}%")

    # Write predictions to CSV file
    with open('/content/drive/MyDrive/submission4.csv', 'w', newline='') as csvfile:
        fieldnames = ['ID', 'is_correct']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()
        for pred in predictions:
            writer.writerow(pred)

    return accuracy

# Run evaluation
accuracy = evaluate_model_on_test_set(model, tokenizer, test_dataset, prompt)
```
